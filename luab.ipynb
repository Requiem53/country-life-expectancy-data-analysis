```python
# Import necessary libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import statsmodels.api as sm
pd.options.display.max_rows = 4000
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
```

```python
# Load the dataset
url = "Cleaned-Life-Exp.csv" 
data = pd.read_csv(url)
```

```python
# Display first few rows of the dataset
print("First few rows of the data:")
print(data.head())

# Basic information about the dataset
print("\nBasic Info about the data:")
print(data.info())

# Statistical summary of the dataset
print("\nStatistical Summary of the data:")
print(data.describe())

# Check for missing values
print("\nMissing values in each column:")
print(data.isnull().sum())

# print(type(data))
# Correlation matrix
# print("\nCorrelation Matrix:")
# corr_matrix = data.corr()
# print(corr_matrix)
```

```python
correlation = data['Schooling'].corr(data['Life expectancy'], method='pearson')  # or 'spearman'
print(f'Correlation between Education and Life Expectancy: {correlation}')
```

```python
# Find what factor has the most impact
dependent_df = data[["Life expectancy"]].copy()
independent_df = data[["Alcohol","percentage expenditure", "Hepatitis B", "Measles", "BMI", "Polio",
                      "Total expenditure", "Diphtheria", "HIV/AIDS", "GDP", "Population", "thinness  1-19 years",
                      "thinness 5-9 years","Income composition of resources","Schooling"]].copy()
# display(independent_df)
independent_df = sm.add_constant(independent_df)

model = sm.OLS(dependent_df, independent_df).fit()

print(model.summary())
```
